```python
import numpy as np
import scipy as sp
import librosa
import soundfile as sf
import matplotlib.pyplot as plt  
import pywt
import sounddevice as sd
```

```python
n = np.linspace(0, 2, 1000)

x1 = 2*np.cos(2*np.pi*70*n) + 3*np.sin(2*np.pi*100*n) + 0.1*np.random.normal(0,2,1000)

plt.plot(x1)
plt.xlabel('n')
plt.ylabel('x1[n]')
```

```python
S1 = librosa.stft(x1, win_length = 40, hop_length = 20)
print(S1.shape)

S2 = librosa.stft(x1, win_length = 40, hop_length = 40)
print(S2.shape)

S3 = librosa.stft(x1, win_length = 160, hop_length = 80)
print(S3.shape)
```

```python
f = np.linspace(0, 500, 1025)
```

```python
t1 = np.linspace(0, 2, 51)
plt.pcolormesh(t1, f , np.abs(S1),shading='auto')
plt.xlabel('Time(sec)')
plt.ylabel('Frequency(Hz)')
```

```python
t2 = np.linspace(0, 2, 26)
plt.pcolormesh(t2, f , np.abs(S2),shading='auto')
plt.xlabel('Time(sec)')
plt.ylabel('Frequency(Hz)')
```

```python
t3 = np.linspace(0, 2, 13)
plt.pcolormesh(t3, f , np.abs(S3),shading='auto')
plt.xlabel('Time(sec)')
plt.ylabel('Frequency(Hz)')
```

```python
s = np.power(2, np.linspace(1, 6, 100))
coefs, freqs = pywt.cwt(x1, s, "cmor3.0-1.0")

t = np.linspace(0,2,1000)
f = freqs * 1000
np.min(f)

plt.pcolormesh(t,f,np.abs(coefs),shading='auto')
plt.xlabel('Time(sec)')
plt.ylabel('Frequency')
```

```python
ns = np.linspace(0,2,1000)

x2 = 1.5*np.cos(2*np.pi*80*ns) + 0.15*np.random.normal(0,2,1000) + \
1.7*(ns-0.725) - (ns-2)

x2[725] += 2
x2[900] += 2

plt.plot(x2)
plt.xlabel('n')
plt.ylabel('x2[n]')
```

```python
S4 = librosa.stft(x2, win_length = 40, hop_length = 20)
print(S2.shape)

t = np.linspace(0, 2, 51)
f = np.linspace(0, 500, 1025)

plt.contour(t, f , np.abs(S4),20)
plt.xlabel('Time(sec)')
plt.ylabel('Frequency')
```

```python
S5 = librosa.stft(x2, win_length = 80, hop_length = 40)
print(S5.shape)

t = np.linspace(0, 2, 26)

plt.contour(t, f , np.abs(S5),20)
plt.xlabel('Time(sec)')
plt.ylabel('Frequency')
```

```python
S6 = librosa.stft(x2, win_length = 160, hop_length = 80)
print(S6.shape)

t = np.linspace(0, 2, 13)

plt.contour(t, f , np.abs(S6),20)
plt.xlabel('Time(sec)')
plt.ylabel('Frequency')
```

```python
s = np.power(2, np.linspace(1, 6, 100))
coefs, freqs = pywt.cwt(data = x2, scales = s, wavelet = "cmor3.0-1.0")

print(coefs.shape)
print(freqs.shape)

t = np.linspace(0,2,1000)
f = freqs*1000

plt.contour(t,f,np.abs(coefs),20)
plt.xlabel('Time(sec)')
plt.ylabel('Frequency')
```
